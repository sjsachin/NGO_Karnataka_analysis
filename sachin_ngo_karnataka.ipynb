```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
data_1=pd.read_csv("Book1.csv")
print(data_1)
```

```python
# making new data frame including total person,illiterate and literate people
data_2=data_1[["Total - Persons","Total - Males","Total - Females","Area Name","Illiterate - Persons","Illiterate - Males","Illiterate - Females","Literate - Persons","Literate - Males","Literate - Females"]]
data_3=data_2.groupby("Area Name").sum()
print(data_3)
```

```python
# ploting total poulation,total male,female population

x=np.arange(len(data_3["Total - Persons"]))
fig=plt.figure(figsize=(30,15))
bar_width=0.25
plt.bar(x,data_3["Total - Persons"],color="g",width=0.25,label="total population")
plt.bar(x+bar_width,data_3["Total - Males"],color="r",width=0.25,label="male population")
plt.bar(x+(bar_width)*2,data_3["Total - Females"],color="b",width=0.25,label="female population")
plt.title("bar garph representing population of different city",size=50)
plt.xlabel("city",size=30)
plt.ylabel("population of city",size=30)
plt.xticks(x+(bar_width),city_list,rotation=90,size=20)
plt.yticks(size=20)
plt.legend(loc=1)
plt.show()
```

```python
#rounding of fraction into four digits followed by converting it into percentage

from decimal import Decimal
lst_1=list()
lst_2=list()
for x,y in zip(data_3["Illiterate - Persons"],data_3["Total - Persons"]):
    k=x/y
    lst_1.append(k)
    
print(lst_1)
print("")
 
for i in lst_1    :
    output=round(i,4)
    illiterate_people=100*(output)
    print(illiterate_people)
    lst_2.append(illiterate_people)  
    #print(lst_2)
data_3["fraction of illiterate person"]=lst_1  
data_3["percentage of illiterate person"]=lst_2
print(data_3.info())
    
   
```

```python
# ploting for total % illiterate people in different city

x=np.arange(len(data_3["percentage of illiterate person"]))
fig=plt.figure(figsize=(30,15))
bar_width=0.25
#plt.bar(x,data_3["fraction of illiterate person"],color="g",width=0.25,label="fraction of illiterate person")
ax=data_3["percentage of illiterate person"].plot(kind="bar",color="g")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),size=20,)
plt.title("bar garph representing Percentage of illiterate persons in different city",size=50)
plt.xlabel("city",size=30)
plt.ylabel("percentage",size=30)
plt.xticks(x+(bar_width),city_list,rotation=90,size=20)
plt.yticks(size=20)
plt.show()
```
